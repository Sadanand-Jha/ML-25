```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
```

```python
train_df = pd.read_csv('train.csv')
```

```python
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
y_train = train_df['Survived']
```

```python
train_df.drop(['Name', 'PassengerId', 'Cabin', 'Ticket'], axis = 1, inplace = True)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
trf1 = make_column_transformer(
    (SimpleImputer(strategy='mean'), [2]),
    (SimpleImputer(strategy='most_frequent'), [6]),
    remainder = 'passthrough')
```

```python
copy_df = train_df
```

```python
trf1.fit_transform(train_df)
```

```python
copy_df = trf1.fit_transform(copy_df)
```

```python
copy_df
```

```python
trf2 = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore', drop='first'), [1, 3]),
    remainder='passthrough'
)
```

```python
copy_df = trf2.fit_transform(copy_df)
```

```python
copy_df
```

```python
trf3 = make_column_transformer(
    (StandardScaler(), [7]),
    remainder='passthrough'
)
```

```python
copy_df = trf3.fit_transform(copy_df)
```

```python
copy_df
```

```python
pipe = make_pipeline(
    (trf1),
    (trf2),
    (trf3)
)
```

```python
pipe.fit(train_df)
```

```python
train_df = pipe.transform(train_df)
```

```python
test_df = pipe.transform(test_df)
```

```python
test_df
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(train_df, y_train)
```

```python
y_pred = dt.predict(test_df)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_test = pd.read_csv('gender_submission.csv')
```

```python
y_test
```

```python
accuracy_score(y_pred, y_test['Survived'])
```

```python
submission = pd.DataFrame({
    'PassengerId': y_test['PassengerId'],  # Replace with your actual test DataFrame
    'Survived': y_pred
})
```

```python
submission.to_csv('submission.csv', index=False)
```
